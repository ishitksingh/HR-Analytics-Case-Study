```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("HR-Employee-Attrition.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition Count")
plt.show()

```

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Income vs Attrition")
plt.show()
```

```python
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title("Age vs Attrition")
plt.show()
```

```python
sns.countplot(x='Department', hue='Attrition', data=df)
plt.xticks(rotation=30)
plt.title("Department-wise Attrition Comparison")
plt.show()
```

```python
sns.countplot(x='WorkLifeBalance', hue='Attrition', data=df)
plt.title("Work Life Balance vs Attrition")
plt.show()
```

Employees with lower income show higher attrition.
Higher salary employees are more likely to stay.
Work-life balance also impacts attrition.
Younger employees leave more than older employees.
